```python
import pandas as pd
```

```python
df= pd.read_csv("heart.csv")
df
```

```python
df1= df[df.Cholesterol<=(3*df.Cholesterol.std()+ df.Cholesterol.mean())]
df1
```

```python
df2=df1[df1.RestingBP<=(3*df1.RestingBP.std()+ df1.RestingBP.mean())]
df2
```

```python
df3=df2[df2.Oldpeak<=(3*df2.Oldpeak.std()+ df2.Oldpeak.mean())]
df3
```

```python
df4= df3.copy()
df4.ST_Slope.replace({'Up':1,'Flat':2,'Down':3}, inplace=True)
df4.ExerciseAngina.replace({'N':0,'Y':1}, inplace=True)
df4.RestingECG.replace({'Normal':1, 'ST':2, 'LVH':3}, inplace=True)
df4.ChestPainType.replace({'ATA':1, 'NAP':2, 'ASY':3, 'TA':4}, inplace=True)
```

```python
df4
```

```python
Dummies=pd.get_dummies(df4.Sex)
Dummies=Dummies.drop(['F'],axis='columns')
merged=pd.concat([df4, Dummies], axis='columns')
X= merged.drop(['HeartDisease','Sex'], axis='columns')
X
```

```python
y=merged.HeartDisease
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
X_scaled= scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X_scaled, y, test_size=0.2, random_state=30)
```

```python
from sklearn.svm import SVC
```

```python
model_svm= SVC()
model_svm.fit(X_train, y_train)
```

```python
model_svm.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc= RandomForestClassifier(n_estimators=40)
rfc.fit(X_train, y_train)
```

```python
rfc.score(X_test, y_test)
```
